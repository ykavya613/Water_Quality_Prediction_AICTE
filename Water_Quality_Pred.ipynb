```python
 import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file (change the path if needed)
file_path = "PB_All_2000_2021.csv"
df = pd.read_csv(file_path, delimiter=';')

# Convert the 'date' column to datetime
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')

# Sort the data by date
df_sorted = df.sort_values(by='date')

# Plot selected water quality parameters
plt.figure(figsize=(14, 10))
parameters = ['NH4', 'O2', 'NO3', 'PO4', 'CL']

for i, param in enumerate(parameters, start=1):
    plt.subplot(3, 2, i)
    plt.plot(df_sorted['date'], df_sorted[param], label=param, color='tab:blue')
    plt.title(f'{param} Over Time')
    plt.xlabel('Date')
    plt.ylabel(param)
    plt.grid(True)
    plt.tight_layout()

plt.suptitle('Water Quality Parameters (2000â€“2021)', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```
